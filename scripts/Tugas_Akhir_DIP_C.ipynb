```python
import re

# IDENTIFIKASI NOISE: EMOJI, URL, MENTION, HASHTAG
# Soal Minggu 3-4: identifikasi noise termasuk emoji berlebihan

def hitung_emoji(text):
    emoji_pattern = re.compile(
        u'[\U0001F600-\U0001F64F'
        u'\U0001F300-\U0001F5FF'
        u'\U0001F680-\U0001F6FF'
        u'\U0001F1E0-\U0001F1FF'
        u'\U00002702-\U000027B0'
        u'\U000024C2-\U0001F251]+',
        flags=re.UNICODE
    )
    return len(emoji_pattern.findall(str(text)))

df['jumlah_emoji'] = df['full_text'].apply(hitung_emoji)

total              = len(df)
tweet_ada_emoji    = (df['jumlah_emoji'] > 0).sum()
total_emoji        = df['jumlah_emoji'].sum()
tweet_ada_url      = df['full_text'].str.contains(r'http\S+|www\.', na=False).sum()
tweet_ada_mention  = df['full_text'].str.contains(r'@\w+', na=False).sum()
tweet_ada_hashtag  = df['full_text'].str.contains(r'#\w+', na=False).sum()
tweet_ada_rt       = df['full_text'].str.lower().str.contains(r'^rt ', na=False).sum()

print('='*55)
print('IDENTIFIKASI NOISE (MINGGU 3-4)')
print('='*55)
print(f'Tweet mengandung emoji    : {tweet_ada_emoji:,} ({tweet_ada_emoji/total*100:.1f}%)')
print(f'Total karakter emoji      : {total_emoji:,}')
print(f'Tweet mengandung URL      : {tweet_ada_url:,} ({tweet_ada_url/total*100:.1f}%)')
print(f'Tweet mengandung mention  : {tweet_ada_mention:,} ({tweet_ada_mention/total*100:.1f}%)')
print(f'Tweet mengandung hashtag  : {tweet_ada_hashtag:,} ({tweet_ada_hashtag/total*100:.1f}%)')
print(f'Tweet retweet (RT)        : {tweet_ada_rt:,} ({tweet_ada_rt/total*100:.1f}%)')
print()
print('Semua jenis noise teridentifikasi')
print('   => Akan dibersihkan pada tahap Minggu 4-6 (Data Cleaning)')
```

# HAPUS NOISE (REGEX CLEANING)

```python
def clean_text(text):
    """Membersihkan teks tweet dari noise."""
    text = str(text)
    text = text.lower()                               # lowercase
    text = re.sub(r'http\S+|www\.\S+', '', text)     # hapus URL
    text = re.sub(r'@\w+', '', text)                 # hapus mention
    text = re.sub(r'#\w+', '', text)                 # hapus hashtag
    text = re.sub(r'RT\s+', '', text)                # hapus RT
    text = re.sub(r'[^a-zA-Z\s]', '', text)         # hapus angka & simbol
    text = re.sub(r'\s+', ' ', text).strip()         # normalisasi spasi
    return text
```

# Apply Cleaning

```python
df['clean'] = df['clean'].apply(clean_text)
```

```python
# Keyword-based labeling
negative_keywords = [
    'gagal bayar', 'galbay', 'teror', 'diteror', 'utang', 'tagihan',
    'debt collector', 'pinjaman online ilegal', 'stres', 'stress',
    'takut', 'bunuh diri', 'menyesal', 'nyesel', 'rugi', 'parah',
    'bunga tinggi', 'ditagih', 'tipu', 'nipu', 'penipuan', 'tertipu',
    'ketipu', 'ditipu', 'susah', 'hancur', 'pusing', 'marah',
    'trauma', 'dikejar', 'ancam', 'jeratan', 'terlilit', 'terjerat',
    'ilegal', 'bahaya', 'berbahaya', 'korban', 'kapok', 'ketakutan',
    'dc lapangan', 'sebar kontak', 'nyebar data', 'permalukan',
    'dipermalukan', 'tidak aman', 'tidak terdaftar', 'jerat utang',
    'minta tolong', 'tolong bantu', 'kesal', 'nyebelin',
]
```

Positif

```python
positive_keywords = [
    'membantu', 'terbantu', 'cepat cair', 'proses cepat', 'solusi',
    'bermanfaat', 'aman', 'terpercaya', 'terdaftar ojk', 'resmi',
    'legal', 'bunga rendah', 'cicilan ringan', 'rekomendasi', 'bagus',
    'senang', 'lega', 'berhasil', 'beruntung', 'bijak', 'cerdas',
    'mudah', 'praktis', 'transparan', 'jelas', 'terjangkau',
    'meringankan', 'menguntungkan', 'puas', 'mantap', 'keren',
]
```

# STEP 3 — BUAT FUNGSI LABELING

```python
def labeling(text):
    """Labeling berbasis keyword."""
    text = str(text).lower()
    for word in negative_keywords:
        if word in text:
            return 'negatif'
    for word in positive_keywords:
        if word in text:
            return 'positif'
    return 'netral'
```

# STEP 4 — APPLY LABELING

```python
df['label'] = df['normalized'].apply(labeling)
```

# MINGGU 13–14: PEMODELAN & PERBANDINGAN MODEL


##  Model yang Dibandingkan

| Model | Alasan Dipilih |
|-------|---------------|
| **Logistic Regression** | Model utama , baseline interpretable |
| **Naive Bayes** | Umum untuk teks, cepat, cocok untuk data kategorikal |
| **SVM (Linear)** | Kuat untuk high-dimensional text features |
| **Random Forest** | Ensemble, tahan overfitting |
| **XGBoost** | Gradient boosting, sering menang di data tabular/teks |

> Semua model menggunakan **TF-IDF** sebagai feature extraction.

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                              f1_score, confusion_matrix, classification_report)

# Model imports
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import SMOTE

# Load data labeled
df = pd.read_csv('dataset_pinjol_labeled.csv')
df = df[df['normalized'].notna() & (df['normalized'].str.strip() != '')]
df = df.reset_index(drop=True)

print(f'Total data: {len(df):,}')
print(df['label'].value_counts())
```

```python
# STEP 1: PERSIAPAN DATA

X = df['normalized']
y = df['label']

# Encode label: negatif=0, netral=1, positif=2
le = LabelEncoder()
y_encoded = le.fit_transform(y)
print('Label classes:', le.classes_)

# Split: 80% train, 20% test (stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print(f'Train: {len(X_train):,} | Test: {len(X_test):,}')

# TF-IDF Vectorizer (shared)
tfidf = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2),    # unigram + bigram
    min_df=2,              # abaikan kata yang muncul < 2x
    sublinear_tf=True      # log normalization
)

X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf  = tfidf.transform(X_test)

print(f'TF-IDF shape: {X_train_tfidf.shape}')
```

```python
# STEP 2: DEFINISI MODEL

models = {
    'Logistic Regression': LogisticRegression(
        max_iter=1000, C=1.0, solver='lbfgs', multi_class='multinomial',
        random_state=42
    ),
    'Naive Bayes': MultinomialNB(
        alpha=0.1  # smoothing parameter
    ),
    'SVM (Linear)': LinearSVC(
        C=1.0, max_iter=2000, random_state=42
    ),
    'Random Forest': RandomForestClassifier(
        n_estimators=100, max_depth=20, random_state=42, n_jobs=-1
    ),
    'XGBoost': XGBClassifier(
        n_estimators=100, max_depth=5, learning_rate=0.1,
        use_label_encoder=False, eval_metric='mlogloss',
        random_state=42, verbosity=0
    ),
}

print('Model siap dijalankan:')
for name in models:
    print(f'  - {name}')
```

```python
# STEP 3: TRAINING & EVALUASI SEMUA MODEL

results = []

for name, model in models.items():
    print(f'\n Training: {name}...')

    # Train
    model.fit(X_train_tfidf, y_train)

    # Predict
    y_pred = model.predict(X_test_tfidf)

    # Metrics
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    rec  = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1   = f1_score(y_test, y_pred, average='weighted', zero_division=0)

    # Cross-validation (5-fold)
    cv_scores = cross_val_score(
        model, X_train_tfidf, y_train, cv=5, scoring='accuracy', n_jobs=-1
    )

    results.append({
        'Model': name,
        'Accuracy': round(acc * 100, 2),
        'Precision': round(prec * 100, 2),
        'Recall': round(rec * 100, 2),
        'F1-Score': round(f1 * 100, 2),
        'CV Mean': round(cv_scores.mean() * 100, 2),
        'CV Std': round(cv_scores.std() * 100, 2),
    })

    print(f'  Accuracy : {acc*100:.2f}%')
    print(f'  F1-Score : {f1*100:.2f}%')
    print(f'  CV Mean  : {cv_scores.mean()*100:.2f}% (±{cv_scores.std()*100:.2f}%)')

print('\n Semua model selesai dilatih!')
```

```python
# STEP 4: TABEL PERBANDINGAN

results_df = pd.DataFrame(results).sort_values('F1-Score', ascending=False)
results_df = results_df.reset_index(drop=True)
results_df.index += 1

print('='*75)
print('TABEL PERBANDINGAN PERFORMA MODEL')
print('='*75)
display(results_df)

# Simpan
results_df.to_csv('hasil_perbandingan_model.csv', index=True)
print('\nTersimpan: hasil_perbandingan_model.csv')
```

```python
# STEP 5: VISUALISASI PERBANDINGAN

metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
model_names = results_df['Model'].values

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

colors = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6']

for i, metric in enumerate(metrics):
    values = results_df[metric].values
    bars = axes[i].barh(model_names, values, color=colors, edgecolor='white', linewidth=0.8)
    axes[i].set_title(f'{metric} (%)', fontsize=12, fontweight='bold')
    axes[i].set_xlim(0, 105)
    axes[i].axvline(x=values.max(), color='red', linestyle='--', alpha=0.4, linewidth=1)
    for bar, val in zip(bars, values):
        axes[i].text(val + 0.5, bar.get_y() + bar.get_height()/2,
                     f'{val:.1f}%', va='center', fontsize=9, fontweight='bold')

plt.suptitle('Perbandingan Performa Model Analisis Sentimen Pinjaman Online',
             fontsize=13, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('perbandingan_model.png', dpi=150, bbox_inches='tight')
plt.show()
print(' Grafik tersimpan: perbandingan_model.png')
```

```python
# STEP 6: CONFUSION MATRIX — LOGISTIC REGRESSION
# (Model utama sesuai judul)

lr_model = models['Logistic Regression']
y_pred_lr = lr_model.predict(X_test_tfidf)

cm = confusion_matrix(y_test, y_pred_lr)
class_names = le.classes_

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Raw counts
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names, ax=axes[0])
axes[0].set_title('Confusion Matrix — Logistic Regression\n(Raw Count)', fontsize=11, fontweight='bold')
axes[0].set_ylabel('Actual', fontsize=10)
axes[0].set_xlabel('Predicted', fontsize=10)

# Normalized
cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='YlOrRd',
            xticklabels=class_names, yticklabels=class_names, ax=axes[1])
axes[1].set_title('Confusion Matrix — Logistic Regression\n(Normalized)', fontsize=11, fontweight='bold')
axes[1].set_ylabel('Actual', fontsize=10)
axes[1].set_xlabel('Predicted', fontsize=10)

plt.tight_layout()
plt.savefig('confusion_matrix_lr.png', dpi=150, bbox_inches='tight')
plt.show()

print('\n Classification Report — Logistic Regression:')
print(classification_report(y_test, y_pred_lr, target_names=class_names))
```

```python
# STEP 7: CONFUSION MATRIX SEMUA MODEL

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, (name, model) in enumerate(models.items()):
    y_pred_i = model.predict(X_test_tfidf)
    cm_i = confusion_matrix(y_test, y_pred_i)
    cm_norm_i = cm_i.astype('float') / cm_i.sum(axis=1)[:, np.newaxis]

    acc_i = accuracy_score(y_test, y_pred_i)
    sns.heatmap(cm_norm_i, annot=True, fmt='.2f', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names, ax=axes[i])
    axes[i].set_title(f'{name}\nAcc: {acc_i*100:.1f}%', fontsize=10, fontweight='bold')
    axes[i].set_ylabel('Actual')
    axes[i].set_xlabel('Predicted')

# Sembunyikan subplot ke-6 (kosong)
axes[-1].set_visible(False)

plt.suptitle('Confusion Matrix Semua Model (Normalized)', fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('confusion_matrix_all.png', dpi=150, bbox_inches='tight')
plt.show()
print(' Tersimpan: confusion_matrix_all.png')
```

```python
# STEP 8: ANALISIS FEATURE IMPORTANCE — Logistic Regression

feature_names = tfidf.get_feature_names_out()

# Koefisien LR untuk setiap kelas
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, class_name in enumerate(class_names):
    coef = lr_model.coef_[i]
    top_indices = np.argsort(coef)[-15:][::-1]
    top_features = [feature_names[j] for j in top_indices]
    top_values = coef[top_indices]

    colors_bar = ['#e74c3c' if class_name == 'negatif' else
                  '#2ecc71' if class_name == 'positif' else '#95a5a6'] * len(top_features)

    axes[i].barh(range(len(top_features)), top_values, color=colors_bar)
    axes[i].set_yticks(range(len(top_features)))
    axes[i].set_yticklabels(top_features, fontsize=9)
    axes[i].set_title(f'Top 15 Fitur — Kelas: {class_name.upper()}',
                      fontsize=11, fontweight='bold')
    axes[i].set_xlabel('Koefisien LR')

plt.suptitle('Feature Importance — Logistic Regression per Kelas Sentimen',
             fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('feature_importance_lr.png', dpi=150, bbox_inches='tight')
plt.show()
print('Tersimpan: feature_importance_lr.png')
```

```python
# STEP 10: PREDIKSI TEKS BARU

import re

def predict_sentiment(text, model=None, vectorizer=None, encoder=None):
    """Prediksi sentimen untuk teks baru."""
    if model is None:
        model = lr_model
    if vectorizer is None:
        vectorizer = tfidf
    if encoder is None:
        encoder = le

    # Preprocessing
    text_clean = re.sub(r'http\S+|@\w+|#\w+|[^a-zA-Z\s]', '', text.lower()).strip()
    text_norm  = ' '.join([slang_dict.get(w, w) for w in text_clean.split()])

    # Vectorize & Predict
    vec  = vectorizer.transform([text_norm])
    pred = model.predict(vec)[0]

    # Probabilitas (jika model mendukung)
    try:
        prob = model.predict_proba(vec)[0]
        prob_dict = {encoder.classes_[i]: f'{p*100:.1f}%' for i, p in enumerate(prob)}
    except:
        prob_dict = None

    label = encoder.inverse_transform([pred])[0]
    return label, prob_dict, text_norm

# Contoh prediksi
test_texts = [
    'pinjol ini sangat membantu saat darurat, proses cair cepat dan bunganya terjangkau',
    'galbay pinjol bikin hidup susah, dc terus telpon dan ancam keluarga gw',
    'ada info pinjaman online terpercaya gak? butuh dana darurat nih',
    'pinjol ilegal banyak banget, hati hati jangan sampai tertipu',
    'baru tau kalo ada pinjol yang terdaftar ojk, lebih aman ya',
]

print(' Contoh Prediksi Sentimen:')
print('='*65)
for text in test_texts:
    label, probs, norm = predict_sentiment(text)
    
    print(f'\n Input : {text[:60]}...' if len(text) > 60 else f'\n Input : {text}')
    print(f'   Norm  : {norm[:60]}')
    print(f'   Hasil : {label.upper()}')
    if probs:
        print(f'   Prob  : {probs}')
```
